```python
import sys
import os
import matplotlib.pyplot as plt

# Ensure the parent directory is in the system path
parent_dir = os.path.abspath("..")
if parent_dir not in sys.path:
    sys.path.insert(0, parent_dir)

from src.utils import find_top_websites
from src.loader import NewsDataLoader

# Initialize the data loader
data_loader = NewsDataLoader()

# Load the news data
news_dataset_path = '../data/data.csv'
news_data = data_loader.load_data(news_dataset_path)

# Find the top websites
top_news_websites = find_top_websites(news_data)
print(top_news_websites)

# Plot the data
plt.figure(figsize=(10,6))
top_news_websites.plot(kind='bar',color='blue')
plt.title('Top websites with the largest count of news articles')
plt.xlabel('Website Domain')
plt.ylabel('No of News Articles')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

```python
import sys
import os
import matplotlib.pyplot as plt

# Ensure the parent directory is in the system path
parent_dir = os.path.abspath("..")
if parent_dir not in sys.path:
    sys.path.insert(0, parent_dir)

from src.utils import find_high_traffic_websites
from src.loader import NewsDataLoader

# Initialize the data loader
data_loader = NewsDataLoader()

# Load the traffic data
traffic_dataset_path = '../data/traffic.csv'
traffic_data = data_loader.load_data(traffic_dataset_path)

# Find the top traffic websites
top_traffic_websites = find_high_traffic_websites(traffic_data)
print(top_traffic_websites)

# Plot the data
plt.figure(figsize=(10,6))
top_traffic_websites.plot(kind='line',marker='o',color='green',linestyle='-')
plt.title('Top websites with the highest traffic')
plt.xlabel('Website')
plt.ylabel('Total IPs')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import sys
import os
import matplotlib.pyplot as plt

# Ensure the parent directory is in the system path
parent_dir = os.path.abspath("..")
if parent_dir not in sys.path:
    sys.path.insert(0, parent_dir)

from src.utils import find_countries_with_most_media
from src.loader import NewsDataLoader

# Initialize the data loader
data_loader = NewsDataLoader()

# Load the domain location data
domain_location_dataset_path = '../data/domains_location.csv'
domain_location_data = data_loader.load_data(domain_location_dataset_path)

# Find the countries with the most media organizations
countries_with_most_media = find_countries_with_most_media(domain_location_data)
print(countries_with_most_media)

# Plot the data
plt.figure(figsize=(8,8))
plt.pie(countries_with_most_media, labels=countries_with_most_media.index, autopct='%1.1f%%', startangle=140)
plt.title('Countries with the highest number of news media organizations')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
